```python
import openslide
import numpy
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
path = 'slide_images\\Tumor_003.tif'
```

```python
slide = openslide.open_slide(path)
```

```python
img = slide.read_region((91600, 140500), 0,
                       (4000, 3000))
```

```python
img.resize((200,200))
```

```python
img_conv = img.convert(mode='P', palette='ADAPTIVE')
img_conv.resize((200,200))
```

```python
adapt_x = numpy.array(img_conv).flatten()
norm_x = numpy.array(img).flatten()
```

```python
plt.hist(norm_x)
```

```python
plt.hist(adapt_x)
```

```python
img_conv.palette
```

```python
imarray = numpy.random.rand(200,200,3) * 255
rand_img = Image.fromarray(imarray.astype('uint8')).convert('RGB')
rand_img
```

```python
rand_img_conv = rand_img.quantize(palette=img_conv)
rand_img_conv
```

```python
adapt_x = numpy.array(rand_img_conv).flatten()
norm_x = numpy.array(rand_img).flatten()
```

```python
plt.hist(norm_x)
```
